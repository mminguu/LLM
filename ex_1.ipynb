```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from konlpy.tag import Okt

# 1ï¸âƒ£ ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
url = "https://drive.google.com/uc?id=1nS74XHctc1WZamEe-sosu2GuLeP7p9jH"
df = pd.read_csv(url)

# 2ï¸âƒ£ ì „ì²´ ë¦¬ë·° ë°ì´í„° ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
reviews = df['review'].astype(str).tolist()

# 3ï¸âƒ£ ë‚´ ë¬¸ì¥
my_review = "ì˜í™” ì²´ì¸ì†Œë§¨ì— ë‚˜ì˜¤ëŠ” ì—¬ì£¼ì¸ê³µì€ ë„ˆë¬´ ì˜ˆì˜ê³ , ì¬ë°Œë‹¤."

# 4ï¸âƒ£ í˜•íƒœì†Œ ë¶„ì„ê¸° (tokenizer)
okt = Okt()
def custom_tokenizer(text):
    return okt.morphs(text, stem=True)

# 5ï¸âƒ£ TF-IDF ë²¡í„°í™”
tfidf = TfidfVectorizer(
    max_features=1000,
    tokenizer=custom_tokenizer
)
tfidf_matrix = tfidf.fit_transform(reviews)

# 6ï¸âƒ£ ë‚´ ë¬¸ì¥ì„ TF-IDFë¡œ ë³€í™˜
query_vec = tfidf.transform([my_review])

# 7ï¸âƒ£ ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ê³„ì‚°
similarities = cosine_similarity(query_vec, tfidf_matrix)[0]

# 8ï¸âƒ£ ìƒìœ„ 5ê°œ ì¸ë±ìŠ¤
top5_idx = similarities.argsort()[::-1][:5]

# 9ï¸âƒ£ ê²°ê³¼ ì¶œë ¥
print("ğŸ“Œ [ê¸°ì¤€ ë¬¸ì¥]:", my_review)
print("="*80)
for rank, idx in enumerate(top5_idx, start=1):
    print(f"{rank}. ìœ ì‚¬ë„: {similarities[idx]:.4f}")
    print(f"ë¦¬ë·°: {reviews[idx]}")
    print("-"*80)
```
